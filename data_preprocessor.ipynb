```python
from transformers import MgpstrProcessor, MgpstrForSceneTextRecognition
from PIL import Image
import cv2
import torch


def obtain_captcha_text(filename):
    # Assuming processor and model are already defined
    processor = MgpstrProcessor.from_pretrained('alibaba-damo/mgp-str-base')
    model = MgpstrForSceneTextRecognition.from_pretrained('alibaba-damo/mgp-str-base')


    # Load image
    # captcha_image_path = "image_file.png"
    captcha_image_path = filename
    image = cv2.imread(captcha_image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB since PIL needs RGB
    image = Image.fromarray(image)

    # Process the image and get pixel values
    pixel_values = processor(images=image, return_tensors="pt").pixel_values

    # Generate model output
    with torch.no_grad():  # Prevent backpropagation since you're not training
        outputs = model(pixel_values)

    # Decode the output to get generated text
    generated_text = processor.batch_decode(outputs.logits)

    output = generated_text['generated_text'][0].upper()
    print(output)

    return output
```

```python
import re
from concurrent.futures import ThreadPoolExecutor

# Function to extract information from LLM output
def extract_ticker_info(llm_output):
    patterns = {
        "Ticker Symbol": r"Ticker Symbol:\s*(\S+)",
        "Exchange": r"Exchange:\s*(.+)",
        "Asset Class": r"Asset Class:\s*(.+)",
        "Currency": r"Currency:\s*(\S+)",
        "Country": r"Country:\s*(.+)",
        "Sector": r"Sector:\s*(.+)",
        "Industry": r"Industry:\s*(.+)",
        "Full Name": r"Full Name:\s*(.+)",
    }
    
    extracted_info = {}
    for key, pattern in patterns.items():
        match = re.search(pattern, llm_output)
        if match:
            extracted_info[key] = match.group(1).strip()
    
    return extracted_info

# Function to process a single ticker
def process_ticker(ticker):
    try:
        llm_output = query_using_groq(user_input=ticker)
        info = extract_ticker_info(llm_output)
        
        return {
            "Ticker Symbol": info.get("Ticker Symbol", ""),
            "Exchange": info.get("Exchange", ""),
            "Asset Class": info.get("Asset Class", ""),
            "Currency": info.get("Currency", ""),
            "Country": info.get("Country", ""),
            "Sector": info.get("Sector", ""),
            "Industry": info.get("Industry", ""),
            "Full Name": info.get("Full Name", ""),
            "Status": True,
        }
    except Exception as e:
        return {"Status": False, "Error": str(e)}

# Parallel processing for all tickers
def process_tickers_in_parallel(files):
    results = []
    with ThreadPoolExecutor() as executor:
        futures = [executor.submit(process_ticker, ticker) for ticker in files]
        for future in futures:
            results.append(future.result())
    return results

# Example usage
# files = ["AAPL", "MSFT", "GOOGL"]  # Replace with your list of tickers
results = process_tickers_in_parallel(files)

# Collate results into separate lists
symbol = []
exchange = []
asset_class = []
currency = []
country = []
sector = []
industry = []
fullname = []

for result in results:
    if result["Status"]:
        symbol.append(result["Ticker Symbol"])
        exchange.append(result["Exchange"])
        asset_class.append(result["Asset Class"])
        currency.append(result["Currency"])
        country.append(result["Country"])
        sector.append(result["Sector"])
        industry.append(result["Industry"])
        fullname.append(result["Full Name"])

# Display or process the results
print("Symbols:", symbol)
print("Exchanges:", exchange)
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from statsmodels.api import OLS, add_constant

# Function to fetch close prices
def fetch_close_prices(tickers, start_date, end_date):
    data = yf.download(tickers, start=start_date, end=end_date)['Adj Close']
    return data

# Detect timeframe
def detect_timeframe(close_prices):
    time_deltas = close_prices.index.to_series().diff().dropna()
    avg_delta = time_deltas.mode().iloc[0]
    return avg_delta

# Compute log returns
def compute_returns(close_prices):
    return np.log(close_prices / close_prices.shift(1))


# Compute rolling beta
def compute_beta_dataframe(close_prices, benchmark, window_days, exclude_benchmark=True):
    timeframe = detect_timeframe(close_prices)
    returns = (compute_returns(close_prices)).dropna()
    benchmark_returns = returns[benchmark]

    if exclude_benchmark: # exclude the benchmark from the dataframe
        asset_returns = returns.drop(columns=[benchmark])
    else:
        asset_returns = returns

    window_periods = int(round(window_days / (timeframe.total_seconds() / 86400)))
    betas = {}

    
    for asset in asset_returns.columns:
        rolling_betas = []
        for i in range(len(benchmark_returns) - window_periods + 1):
            window_benchmark = benchmark_returns.iloc[i:i + window_periods]
            window_asset = asset_returns[asset].iloc[i:i + window_periods]
            X = add_constant(window_benchmark)
            model = OLS(window_asset, X).fit()
            rolling_betas.append(model.params[benchmark])
        padding = [np.nan] * (window_periods - 1)
        betas[asset] = padding + rolling_betas
    return (pd.DataFrame(betas, index=returns.index)).dropna()


# Compute excess returns
def compute_excess_returns_dataframe(close_prices, benchmark):
    returns = compute_returns(close_prices)
    benchmark_returns = returns[benchmark]
    excess_returns = returns.sub(benchmark_returns, axis=0)
    return excess_returns.drop(columns=[benchmark])

# Main script
if __name__ == "__main__":
    # Define tickers and time period
    stock_tickers = [
        "AAPL", "MSFT", "AMZN", "GOOGL", "TSLA", "META", "NVDA", "JPM", "JNJ", "UNH",
        "V", "PG", "HD", "MA", "DIS", "PFE", "KO", "PEP", "ABBV", "XOM",
        "CSCO", "WMT", "CVX", "ADBE", "NFLX", "NKE", "MRK", "BAC", "INTC", "T",
        "WFC", "MDT", "COST", "VZ", "CRM", "AMGN", "TXN", "QCOM", "ACN", "IBM",
        "ORCL", "TMUS", "LLY", "HON", "LOW", "BA", "MMM", "AVGO", "UPS", "PM"
    ]
    benchmark_ticker = "SPY"
    tickers = stock_tickers + [benchmark_ticker]

    # Fetch close prices
    close_prices = fetch_close_prices(tickers, start_date="2023-01-01", end_date="2023-12-31")

    # Compute rolling beta
    rolling_beta_df = compute_beta_dataframe(close_prices, benchmark="SPY", window_days=30)
    print("Rolling Beta DataFrame:")
    print(rolling_beta_df)

    # Compute excess returns
    excess_returns_df = compute_excess_returns_dataframe(close_prices, benchmark="SPY")
    print("Excess Returns DataFrame:")
    print(excess_returns_df)
```
